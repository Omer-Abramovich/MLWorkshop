```python
import xlrd as xlrd
import torch as torch
import cv2 as cv
import numpy as np
from pathlib import Path
```

```python
def excelToTensor(excelPath,infantOrParent):
    tagData = xlrd.open_workbook(excelPath).sheet_by_index(0)
    file_length=tagData.nrows

    tensorAllVideos = []
    tensorOneVideo = []
    array=[]
    subjectNum = tagData.cell_value(1,16)
    
    for i in range(1,tagData.nrows):
        numberOfReturns=tagData.cell_value(i,0)
        subjectNumTemp=tagData.cell_value(i,16)
        
        if(subjectNumTemp > subjectNum):
            print("subjectNum " + str(subjectNum))
            #tensorAllVideos.append(torch.FloatTensor(tensorOneVideo))
            torch.save(torch.LongTensor(tensorOneVideo), excelPath[:-11] + "Subject_"+str((int)((subjectNum+1)/2))+"/"+str((int)(((subjectNum+1)%2)*3 +3)) +"m/"+infantOrParent+"/Targets.pth")
            subjectNum=subjectNumTemp
            tensorOneVideo = []
        
        for j in range(int(numberOfReturns)):
            for k in range(1,16):
                array.append(tagData.cell_value(i,k))
            tensorOneVideo.append(array)
            array=[]
            
    subjectNum=subjectNumTemp
    torch.save(torch.LongTensor(tensorOneVideo), excelPath[:-11] + "Subject_"+str((int)((subjectNum+1)/2))+"/"+str((int)(((subjectNum+1)%2)*3 +3)) +"m/"+infantOrParent+"/Targets.pth")
    tensorOneVideo = []
```

```python
for path in Path('/data/ml_workshop_processed').glob('**/*Infant.xlsx'):
    print (str(path))
    print("")
    excelToTensor(str(path),"Infant")
    

for path in Path('/data/ml_workshop_processed').glob('**/*Parent.xlsx'):
    excelToTensor(path,"Parent")


```
